# Création par étapes (de la Table 1)

Créons les tables intermédiaires, d'abord pour les variables catégorielles.

```python
a = 3
```

```python
df.filter(
    pl.col('ARMCD') == 'CLON',
).group_by(
    pl.col('SEX'),
    maintain_order=True,
).len(
)
```

```python
df.filter(
    pl.col('ARMCD') == 'CLON',
).group_by(
    pl.col('SEX'),
    maintain_order=True,
).len(
).with_columns(
    percentage=pl.col('len') / pl.sum('len'),
)
```

```python
def cat_table(df, arm, cat_var):
    t = df.filter(
        pl.col('ARMCD') == arm,
    ).group_by(
        pl.col(cat_var),
        maintain_order=True,
    ).len(
    ).with_columns(
        percentage=pl.col('len') / pl.sum('len'),
    )
    return t
```

```python
m_list = []
```

```python
type(m_list)
```

```python
bupn_cat = [cat_table(df, arm='BUPNAL', cat_var=cat) for cat in categorical]
```

```python
clon_cat = [cat_table(df, arm='CLON', cat_var=cat) for cat in categorical]
```

```python
clon_cat
```

Et maintenant pour la variable numérique continue.

```python
effectifs = df.group_by(
    pl.col('ARMCD'),
    maintain_order=True,
).len(
).with_columns(
    percentage=pl.col('len') / pl.sum('len'),
)

effectifs
```

```python
cont = df.group_by(
    pl.col('ARMCD'),
    maintain_order=True,
).agg(
    mean_age=pl.col('AGE').mean(),
    std_age=pl.col('AGE').std(),
    med_age=pl.col('AGE').median(),
)

cont
```

```python
cont.columns
```
